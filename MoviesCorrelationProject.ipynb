```python
# import libraries

import pandas as pd
import seaborn as sn
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use("ggplot")
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)  # Adjusts the configuration of the plots we will create


# read in the data

df = pd.read_csv('/Users/owenma/Downloads/movies.csv')
```

```python
# look at the data

df.head()  # first 5 rows
```

```python
# check if there is any missing data

for col in df.columns:
    percent_missing = np.mean(df[col].isnull())
    print("{} - {}%".format(col,percent_missing))
```

```python
# check data types of each column

df.dtypes
```

```python
# sort the dateset by gross column in desceding order

df.sort_values(by=['gross'], inplace=False, ascending=False).head()
```

```python
# display all rows of a dataset

pd.set_option('display.max_rows', None)
```

```python
# remove any duplicates of the column 'company'

df['company'].drop_duplicates().sort_values(ascending=False).
```

### check columns for high correlation with gross column

### hypothesis: budget and company columns are highly correlated with gross

```python
# Scatter plot with budget vs gross

plt.scatter(x = df['budget'], y=df['gross'])

plt.title('Budget vs Gross Earnings')
plt.xlabel('Budget for film')
plt.ylabel('Gross Earnings')

plt.show()
```

```python
# plot budget vs gross using seaborn

sn.regplot(x='budget', y='gross', data=df, scatter_kws={"color": "red"}, line_kws={"color":"blue"})
```

```python
# check for correlation

df.corr(method='pearson')  #pearson(defalut), kendall, spearman
```

### Outcome: High correlation between budget and gross

```python
# visualize correlation

correlation_matrix = df.corr(method='pearson')

sn.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')

plt.show()
```

```python
# Check for company column

df.head()
```

```python
df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized.head()
```

```python
# visualize correlation again

correlation_matrix = df_numerized.corr(method='pearson')

sn.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')

plt.show()
```

```python
corrlation_matrix = df_numerized.corr()

corr_pairs = corrlation_matrix.unstack()

corr_pairs.head()
```

```python
sorted_pairs = corr_pairs.sort_values()

sorted_pairs.head()
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.5]

high_corr.head()
```
